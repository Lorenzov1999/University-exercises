2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
import numpy as np

u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

#outer function
outermatrix=np.outer(u, v, out=None)
print("with outer")
print(outermatrix)

#nested loop
loopmatrix=np.zeros((4,4))
for i in range (len(u)):
    for j in range (len(v)):
        loopmatrix[i,j]=u[i]*v[j]
print("nested loop")
print(loopmatrix)

#broadcasting operations
broadcastmatrix=u*v.reshape(4,1)
print("broadcasting operations")
print(broadcastmatrix)
````

````python
import matplotlib.pyplot as plt
import numpy as np

def reverser(arrr):
    print("lenarr:\n",len(arrr))
    temporary=np.zeros(len(arrr))
    for j in range (0,len(arrr)):
        temporary[j]=float(arrr[j])
    for k in range (0,100):
        arrr[k]=temporary[len(arrr)-1-k]
    return arrr

def piforther(x):
    delta=abs(np.sin(x)-np.cos(x))
    if delta<=0.1:
        return 1
    else:
        return 0
    
arr=np.linspace(0, 2*np.pi, 100)
print("arr:\n",arr)

print("tenth elemnts:")
for i in range (1,11):
    print(arr[i*10-1])
    
print("reversed arr:\n",reverser(arr))

print("the following elemtents have similar values of sin and cos")
closenumbs=[]
for y in arr:
    if piforther(y)==1:
        print(y)
        print("same element in degrees:",y*180/np.pi)
        closenumbs.append(y)

plt.plot(arr, np.sin(arr))
plt.plot(arr, np.cos(arr))
print("The two plots are close when you see them close(?)")
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import numpy as np
def mult(x):
    multiples=[]
    for i in range (2,x):
        for j in range (2,round(x/2)):
            if i*j not in multiples and i*j<=x:
                multiples.append(i*j)
    if x not in multiples and x>1:
        return True
    else:
        return False

N = 100
a = np.arange(N)
mask=[]
for i in a:
    mask.append(mult(a[i]))
primes=a[mask]
print(primes)
````

````python
def Eratostene(x):
    lst = np.array([y for y in np.arange(x)])
    for i in range(2,x):
        if ( lst[i]!=-1 ):
            k = i
            for j in range(2,int(x/i)):
                if ( k+i<x ):
                    lst[k+i]=-1
                    k = k+i
    lst[0]=-1
    lst[1]=-1
    return lst[lst!=-1]
    
print(Eratostene(100)) 
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
import matplotlib.pyplot as plt
import numpy.random as npr


steps=np.zeros((1000,201))
for j in range (1,len(steps[1,:])):#0-200
    for k in range (0,len(steps[:,1])):#0-1000
        sign=npr.rand(1)*2-1
        if sign<=0:
            steps[k,j]=1
        else:
            steps[k,j]=-1      
print ("steps table\n",steps)

distances=np.zeros((len(steps[:,1]),len(steps[1,:])))
for i in range (1,len(steps[1,:])):
    distances[:,i]=distances[:,i-1]+steps[:,i]
print("distances table\n",distances)

squaredistances=distances**2

meansquares=np.zeros(len(steps[1,:]))
for l in range (0,len(steps[1,:])):
    meansquares[l]=sum(squaredistances[:,l])/len(squaredistances[:,l])

meandistances=np.sqrt(meansquares)
print("mean_distances(t)\n",meandistances)

plt.plot([i for i in range(0,201)],meandistances)
````

Do you feel there is some evident correlation here? [Studies](https://www.enr.gov.nt.ca/en/services/lynx/lynx-snowshoe-hare-cycle) tend to believe so.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#loading, creating arrays
populations = np.loadtxt('populations.txt')
print("The dataset is: \n", populations)
year, hares, lynxes, carrots = populations.T

#plotting
plt.axes([0.2, 0.1, 0.5, 0.8]) 
plt.plot(year, hares, year, lynxes, year, carrots) 
plt.legend(('Hare', 'Lynx', 'Carrot'), loc=(1.05, 0.5)) 

#statistical properties
print("The dominant species ordered per year are:")
for i in range (len(populations[:,0])):
    if np.argmax(populations[i,1:])==0:
        specie="Hare"
    if np.argmax(populations[i,1:])==1:
        specie="Lynx"
    if np.argmax(populations[i,1:])==2:
        specie="Carrot"
    print (round(populations[i,0]),"'s dominant specie:", specie)
print("The means and stds are:")
print(f" - Hares: Mean = {populations[1,:].mean()}, StdDev = {populations[1,:].std()}")
print(f" - Lynxes: Mean = {populations[2,:].mean()}, StdDev = {populations[2,:].std()}")
print(f" - Carrots: Mean = {populations[3,:].mean()}, StdDev = {populations[3,:].std()}")
````

It would make sense if the correlations were between the first derivative of a specie and the value of the other species, so we'll look in that direction.

````python
derivatives=np.zeros((21,4))
for i in range (20):
    for j in range (4):
        derivatives[i,j]=populations[i+1,j]-populations[i,j]
print (derivatives)
````

````python
plt.plot(year, populations[:,1], year, derivatives[:,2])
plt.legend(('Hare', 'deltaLynx'), loc=(1.05, 0.5)) 
np.corrcoef(populations[:,1], derivatives[:,2])
````

````python
plt.plot(year, populations[:,1], year, derivatives[:,3])
plt.legend(('Hare', 'deltaCarrot'), loc=(1.05, 0.5)) 
np.corrcoef(populations[:,1], derivatives[:,3])
````

````python
plt.plot(year, populations[:,2], year, derivatives[:,3])
plt.legend(('Lynx', 'deltaCarrot'), loc=(1.05, 0.5)) 
np.corrcoef(populations[:,2], derivatives[:,3])
````

````python
plt.plot(year, populations[:,2], year, derivatives[:,1])
plt.legend(('Lynx', 'deltaHare'), loc=(1.05, 0.5)) 
np.corrcoef(populations[:,2], derivatives[:,1])
````

````python
plt.plot(year, populations[:,3], year, derivatives[:,1])
plt.legend(('Carrot', 'deltaHare'), loc=(1.05, 0.5)) 
np.corrcoef(populations[:,3], derivatives[:,1])
````

````python
plt.plot(year, populations[:,3], year, derivatives[:,2])
plt.legend(('Carrot', 'deltaLynx'), loc=(1.05, 0.5))
np.corrcoef(populations[:,3], derivatives[:,2])
````
